```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    "Age": [23, 45, 31, 35, 52, 46, 28, 40],
    "BMI": [22.5, 27.8, 31.2, 24.5, 29.0, 33.5, 26.1, 30.2],
    "Charges": [2000, 5000, 7000, 3000, 9000, 11000, 4000, 8000]
}

df = pd.DataFrame(data)

print(df)
```

```python
print(df.shape)

print(df.columns)

print(df.describe())
```

```python
average = df["Charges"].mean()

print("Average Charge:", average)
```

```python
maximum = df["Charges"].max()

print("Maximum Charge:", maximum)
```

```python
plt.plot(df["Age"], df["Charges"], marker='o')

plt.title("Age vs Insurance Charges")

plt.xlabel("Age")

plt.ylabel("Charges")

plt.show()
```

```python
plt.bar(df["Age"], df["BMI"])

plt.title("Age vs BMI")

plt.xlabel("Age")

plt.ylabel("BMI")

plt.show()
```
